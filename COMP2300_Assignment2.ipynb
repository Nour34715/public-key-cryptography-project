## Question 3(a)

```python
import hashlib

h = "aa383eeb2bb5fd5d16d5cef91b18749fc5c03431b9335de54d00151ec80dff94"
s = "48713406"
trans = "Pay 10 bitcoins to " + s

r = 1

while True:
    data = h + trans + str(r)
    hash_val = hashlib.sha256(data.encode()).hexdigest()
    
    if hash_val.startswith("0000") and not hash_val.startswith("00000"):
        print("r =", r)
        print("hash =", hash_val)
        break
    
    r += 1
```

### Explanation

I implemented a hash puzzle using the SHA-256 function. I searched for a nonce r such that the resulting hash has exactly four leading zeros in its hexadecimal representation. I incremented r starting from 1 until the condition was satisfied.

## Question 3(b)

```python
total = 0

h = "aa383eeb2bb5fd5d16d5cef91b18749fc5c03431b9335de54d00151ec80dff94"
s = 48713406

for i in range(1000):
    s_val = s + i
    trans = "Pay 10 bitcoins to " + str(s_val)
    
    r = 1
    while True:
        data = h + trans + str(r)
        hash_val = hashlib.sha256(data.encode()).hexdigest()
        
        if hash_val.startswith("0000") and not hash_val.startswith("00000"):
            total += r
            break
        
        r += 1

avg = total / 1000
print("Average r =", avg)
```

### Explanation

I repeated the hash puzzle 1000 times, starting from my student ID and increasing it by 1 each time up to s + 999. For each value of s, I searched for the first nonce r that produced a SHA-256 hash with exactly four leading zeros in hexadecimal.

The average value of r I obtained was 67063.878. This is close to the expected value of 16^4 = 65536, because each hexadecimal digit has 16 possible values, so the probability of obtaining four leading zeros is 1/16^4. Therefore, on average, about 65536 attempts are expected.
